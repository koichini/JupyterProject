```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time


start = time.time()
url = "https://l-formula.com/words1"

r = requests.get(url)
# if r.status_code != requests.codes.ok:
#     print(f'error: not found{url}')
#     return

soup = BeautifulSoup(r.content, 'html.parser')

word_list_row = [i.get_text() for tr in soup.find_all('tr') for i in tr.find_all('td') ]
del word_list_row[0:4]

word_list = []
length = len(word_list_row)
n = 0
s = 4

for i in word_list_row:
    word_list.append(word_list_row[n:n+s:1])
    n += s
    if n >= length:
        break

columns = ['id', 'word', 'meaning', 'pronunciation']
df1 = pd.DataFrame(word_list, columns=columns)
df = df1.set_index('id')

df
```
